Fit a linear regression and plot the cost function for the *gas consumption dataset with tax, income, highway, drivers and gas*.

$$ J = \frac{1}{m}\sum_{i=1}^m (y_i - \theta_0 - \theta_i x_i)^2 $$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'./dataset/gas_consumption.csv', names = ['tax', 'income', 'highway', 'drivers','gas'])
df
```

```python
# initialise the linear regressor
lr = LinearRegression()

# fit
# reg drivers on gas
lr.fit(df[['drivers']], df['gas']) # fit(X,y)

lr.intercept_, lr.coef_
```
